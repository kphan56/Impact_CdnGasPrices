# Cleaning Data

```python
gasPrices['Date'].describe()
```

The data includes the first day of the month each time. We only need the month and year

```python
gasPrices['Date'] = pd.to_datetime(gasPrices['Date'])
gasPrices['Date'] = gasPrices['Date'].dt.strftime('%m/%Y')
print(gasPrices)
```

There exists some NaN data in the last rows of the Petroleum table. The values by default are float but none of the values are deimcals so it's cleaner to convert them to int

```python
petroleumImports.copy()
petroleumImports = petroleumImports.dropna(axis = 0, ignore_index = False)
petroleumImports['Refined Petroleum'] = petroleumImports['Refined Petroleum'].apply(lambda x: int(x))
petroleumImports['Crude Oil'] = petroleumImports['Crude Oil'].apply(lambda x: int(x))
petroleumImports
```

# Extracting Data

The gas prices are split into three categories per date: Base, tax, and total. To avoid showing all three, they must be properly indexed to not include them all in every table

```python
gas_base = gasPrices[0::3]
gas_base
```

```python
gas_tax = gasPrices[1::3]
gas_tax
```

```python
gas_total = gasPrices[2::3]
gas_total
```

# Plotting Data

Overview of Canadian gas price totals from a few Canadian major cities

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import CrosshairTool, HoverTool, ColumnDataSource

output_notebook()

monthYear = pd.to_datetime(gas_total['Date'], format='%m/%Y').tolist()

gasTotalsVisual = figure(title="Average Gas Prices in Canadian major cities", x_axis_label="Month and Year", x_axis_type="datetime", y_axis_label="Gas Price (c/L)", width=1500, height=950)

cities = ['Toronto', 'Ottawa', 'Thunder Bay', "St. John's, Newfoundland", 'Charlottetown', 'Halifax', 'Montreal', 'Winnipeg', 'Regina', 'Calgary', 'Vancouver']
colors = ['#a056b5', '#baa432', '#c50000', '#40e654', '#761ccc', '#0a9987', '#df12a8', '#561b4c', '#777cab', '#ef1010', '#643bbd', '#00bade']

for city, color in zip(cities, colors):
    source = ColumnDataSource(data={
        'Date': monthYear,
        'Price': gas_total[city],
        'City': [city] * len(monthYear)
    })

    gasTotalsVisual.line(x='Date', y='Price',legend_label=city, line_width=2, color=color, source=source)

gasTotalsVisual.add_tools(HoverTool(
    tooltips=[
        ('City', '@City'),
        ('Date', '@Date{%b. %Y}'),
        ('Gas Price', '@Price{0.2f} ¢/L')
    ],
    formatters={
        '@Date': 'datetime'
    },
))

gasTotalsVisual.add_tools(CrosshairTool())
gasTotalsVisual.legend.location = 'top_left'
gasTotalsVisual.legend.click_policy = 'hide'

show(gasTotalsVisual)
```

Overview of the base Canadian gas from a few Canadian major cities including the imports of Petroleum and Crude Oil

```python
from bokeh.layouts import column, row

output_notebook()

monthYear = pd.to_datetime(gas_base['Date'], format='%m/%Y').tolist()

gasBaseVisual = figure(title="Base Gas Prices in Canadian major cities", x_axis_label="Month and Year", x_axis_type="datetime", y_axis_label="Base Gas Price (c/L)", width=1500, height=950)

cities = ['Toronto', 'Ottawa', 'Thunder Bay', "St. John's, Newfoundland", 'Charlottetown', 'Halifax', 'Montreal', 'Winnipeg', 'Regina', 'Calgary', 'Vancouver']
colors = ['#a056b5', '#baa432', '#c50000', '#40e654', '#761ccc', '#0a9987', '#df12a8', '#561b4c', '#777cab', '#ef1010', '#643bbd', '#00bade']

for city, color in zip(cities, colors):
    source = ColumnDataSource(data={
        'Date': monthYear,
        'Price': gas_base[city],
        'City': [city] * len(monthYear)
    })

    gasBaseVisual.line(x='Date', y='Price',legend_label=city, line_width=2, color=color, source=source)

gasBaseVisual.add_tools(HoverTool(
    tooltips=[
        ('City', '@City'),
        ('Date', '@Date{%b. %Y}'),
        ('Base Gas Price', '@Price{0.2f} ¢/L')
    ],
    formatters={
        '@Date': 'datetime'
    },
))

gasBaseVisual.add_tools(CrosshairTool())
gasBaseVisual.legend.location = 'top_left'
gasBaseVisual.legend.click_policy = 'hide'

#Petroleum Table
petroleumCrudeOilDates = pd.to_datetime(petroleumImports['Date'], format='%B %Y').tolist()

petroleumImportTotal = petroleumImports['Refined Petroleum'].tolist()

petroleumImportsVisual = figure(title="Petroleum Canadian Imports", x_axis_label='Date', y_axis_label='$ (million)', x_axis_type='datetime', height=500)

source = ColumnDataSource(data={
        'PetroleumDate': petroleumCrudeOilDates,
        'PetroleumImports': petroleumImportTotal
})

petroleumImportsVisual.add_tools(HoverTool(
    tooltips=[
        ('Date', '@PetroleumDate{%b. %Y}'),
        ('Petroleum Imports', '$@PetroleumImports million')
    ],
    formatters={
        '@PetroleumDate': 'datetime'
    },
))

petroleumImportsVisual.vbar(x='PetroleumDate', top='PetroleumImports', width=15*24*60*60*1000, color='#000000', source=source)

#Crude Oil Table
crudeOilImportTotal = petroleumImports['Crude Oil'].tolist()
crudeOilImportsVisual = figure(title="Crude Oil Canadian Imports", x_axis_label='Date', y_axis_label='$ (million)', x_axis_type='datetime', height=500)

source = ColumnDataSource(data={
        'CrudeOilDate': petroleumCrudeOilDates,
        'CrudeOilImports': crudeOilImportTotal
})

crudeOilImportsVisual.add_tools(HoverTool(
    tooltips=[
        ('Date', '@CrudeOilDate{%b. %Y}'),
        ('Crude Oil Imports', '$@CrudeOilImports million')
    ],
    formatters={
        '@CrudeOilDate': 'datetime'
    },
))

crudeOilImportsVisual.vbar(x='CrudeOilDate', top='CrudeOilImports', width=15*24*60*60*1000, color='#000001', source=source)

show(column(gasBaseVisual, row(petroleumImportsVisual, crudeOilImportsVisual)))
```

Finding the mean gas price of each city then comparing them to the prices during COVID

```python
gas_total = gas_total.copy()
gas_total['Date'] = pd.to_datetime(gas_total['Date'], format='%m/%Y')

def extractCovidGasPrices(df):
  return gas_total[(gas_total['Date'] > pd.Timestamp('2020-01-01')) & (gas_total['Date'] < pd.Timestamp('2020-06-01'))]

covid_gas = extractCovidGasPrices(gas_total)
print(covid_gas)

cities = ['Toronto', 'Ottawa', 'Thunder Bay', "St. John's, Newfoundland", 'Charlottetown', 'Halifax', 'Montreal', 'Winnipeg', 'Regina', 'Calgary', 'Vancouver']

mean_value={}
for city in cities:
    mean_value[city] = gas_total[city].mean()
    print(f"Mean value for {city}: {mean_value[city]}")
```

```python
output_notebook()

monthYear = pd.to_datetime(covid_gas['Date']).tolist()

cities = ['Toronto', 'Ottawa', 'Thunder Bay', "St. John's, Newfoundland", 'Charlottetown', 'Halifax', 'Montreal', 'Winnipeg', 'Regina', 'Calgary', 'Vancouver']
colors = ['#a056b5', '#baa432', '#c50000', '#40e654', '#761ccc', '#0a9987', '#df12a8', '#561b4c', '#777cab', '#ef1010', '#643bbd', '#00bade']

covidGasVisual = figure(title="Mean Gas Prices vs. Covid Gas Prices", x_axis_label="Impacted Covid 2020 months", x_axis_type="datetime", y_axis_label="Gas Price (c/L)", width=1500, height=950)

for city, color in zip(cities, colors):
    means = mean_value[city]
    size = len(monthYear)
    mean_line = [means] * size
    source = ColumnDataSource(data={
        'Date': monthYear,
        'Price': covid_gas[city],
        'Mean': mean_line,
        'City': [city] * size
    })
    covidGasVisual.line(x='Date', y='Price',legend_label=city, line_width=2, color=color, source=source)
    covidGasVisual.line(x='Date', y='Mean', legend_label=f"{city} Mean", line_dash='dashed', color=color, source=source)

covidGasVisual.add_tools(HoverTool(
  tooltips=[
      ('City', '@City'),
      ('Date', '@Date{%b}'),
      ('Gas Price', '@Price{0.2f} ¢/L')
  ],
  formatters={
      '@Date': 'datetime'
  },
))

covidGasVisual.add_tools(CrosshairTool())
covidGasVisual.legend.location = 'top_right'
covidGasVisual.legend.click_policy = 'hide'

show(covidGasVisual)
```
